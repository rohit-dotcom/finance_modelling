#### Volatile stock= Bitcoin VS Non Volatile stock= Ford Motors
We need to see how different are log return and arithmetic return of a volatile stock and a non volatile stock.

```python
#import the required libraies
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import yfinance as yfin
import datetime
yfin.pdr_override()
```

```python
#get the bitcoin adjusted closing prices for the last year
start = datetime.date(2021, 12, 25)
end = datetime.date(2023, 12, 28)

prices_df=web.DataReader(["BTC-USD","SPY"],start=start,end=end)["Adj Close"]
```

```python
prices_df['SPY_pct_returns']=prices_df['SPY'].pct_change()
prices_df['SPY_log_returns']=np.log(prices_df['SPY']/prices_df['SPY'].shift(1))
prices_df['BTC_pct_returns']=prices_df['BTC-USD'].pct_change()
prices_df['BTC_log_returns']=np.log(prices_df['BTC-USD']/prices_df['BTC-USD'].shift(1))
```

```python
prices_df[['SPY_pct_returns','SPY_log_returns',]].plot(kind='kde',title="S&P 500 index log vs pct returns for last 2 years")
```

```python
prices_df[['BTC_log_returns','BTC_pct_returns']].plot(kind='kde',title="Bitcoin log vs pct returns for last 2 years")
```
